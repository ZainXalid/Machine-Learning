```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import levenberg_marquardt as lm
from sklearn.model_selection import train_test_split


# Load dataset from CSV
dataset_path = "dataset.csv"
df = pd.read_csv(dataset_path)

# Last column is the target variable 
result_column = df.columns[-1]

# features_columns = df.columns[:-1]
features_columns = df.columns[0]

# x = df[features_columns].values
x = df[features_columns].values.reshape(-1, 1)
y = df[result_column].values.reshape(-1, 1)

#Robust Scaling 0.80 / 0.72
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
x_normalized = scaler.fit_transform(x)

#Power Transformations (Box-Cox or Yeo-Johnson): 0.54 / 0.73
# from sklearn.preprocessing import PowerTransformer

# scaler = PowerTransformer(method='yeo-johnson')  # or 'box-cox'
# x_normalized = scaler.fit_transform(x)

### works well for the below normalizations
# Normalize the input features 0.56 / 0.72
# from sklearn.preprocessing import StandardScaler

# scaler = StandardScaler()
# x_normalized = scaler.fit_transform(x)

# MinMax normalization 0.73 / 0.72   (1)
# from sklearn.preprocessing import MinMaxScaler
# scaler = MinMaxScaler()
# x_normalized = scaler.fit_transform(x)

#Quantile Transformation: 0.52 / 0.73 for uniform  and  6.45 / 6.97 for normal    (2)
# from sklearn.preprocessing import QuantileTransformer

# scaler = QuantileTransformer(output_distribution='uniform')  # or 'normal'
# x_normalized = scaler.fit_transform(x)


# Split the dataset into training, validation, and test sets
x_train, x_temp, y_train, y_temp = train_test_split(x_normalized, y, test_size=0.1, random_state=42)
x_val, x_test, y_val, y_test = train_test_split(x_temp, y_temp, test_size=0.5, random_state=42)

# # Define the model
# #     tf.keras.layers.BatchNormalization(),

# model = tf.keras.Sequential([
#     tf.keras.layers.Dense(20, activation='tanh', input_shape=(x_train.shape[1],)),
#     tf.keras.layers.Dense(40, activation='elu'),
#     tf.keras.layers.Dense(20, activation='elu'),
#     tf.keras.layers.Dense(1, activation='linear')])

model = tf.keras.Sequential([
    tf.keras.layers.Dense(40, activation='tanh', input_shape=(x_train.shape[1],)),
    tf.keras.layers.Dense(1, activation='linear')])

model.summary()

# model.compile(
#     optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
#     loss=tf.keras.losses.MeanSquaredError()) #, metrics=['mse']

model_wrapper = lm.ModelWrapper(
    tf.keras.models.clone_model(model))

model_wrapper.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=1),
    loss=lm.MeanSquaredError()) #, metrics=['mse']

# Convert data to TensorFlow Datasets
train_dataset = tf.data.Dataset.from_tensor_slices((x_train, y_train))
train_dataset = train_dataset.shuffle(len(x_train))
train_dataset = train_dataset.batch(1000).cache()
train_dataset = train_dataset.prefetch(tf.data.experimental.AUTOTUNE)

val_dataset = tf.data.Dataset.from_tensor_slices((x_val, y_val))
val_dataset = val_dataset.batch(1000).cache()
val_dataset = val_dataset.prefetch(tf.data.experimental.AUTOTUNE)

# Test dataset
test_dataset = tf.data.Dataset.from_tensor_slices((x_test, y_test))
test_dataset = test_dataset.batch(1000).cache()
test_dataset = test_dataset.prefetch(tf.data.experimental.AUTOTUNE)

# Lists to store metrics for plotting
lm_train_loss = []
lm_val_loss = []
test_losses_lm = []


# Training using Levenberg-Marquardt
print("\n_________________________________________________________________")
print("Train using Levenberg-Marquardt")
t2_start = time.perf_counter()

for epoch in range(150):
    history_lm = model_wrapper.fit(train_dataset, epochs=1, validation_data=val_dataset)
    
    lm_train_loss.append(history_lm.history['loss'][0])
    lm_val_loss.append(history_lm.history['val_loss'][0])
    
    # Evaluate on the test set at each epoch
    test_loss_lm = model_wrapper.evaluate(test_dataset)
    test_losses_lm.append(test_loss_lm)

t2_stop = time.perf_counter()
print("Elapsed time: ", t2_stop - t2_start)

# Plot Mean Squared Error
sns.set(style="white")
plt.figure(figsize=(14, 12))


# Levenberg-Marquardt Optimizer
plt.subplot(2, 2, 2)
plt.plot(lm_train_loss, label='Train')
plt.plot(lm_val_loss, label='Val')
plt.plot(test_losses_lm, label='Test')
plt.title('Mean Squared Error - Levenberg-Marquardt')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Evaluate on the test set
test_loss = model_wrapper.evaluate(x_test, y_test)
print(f'LM Test Loss: {test_loss}')
# Levenberg-Marquardt Optimizer
sns.set(style="white")

plt.subplot(1, 1, 1)
plt.plot(lm_train_loss, label='Train', color='blue')
plt.plot(lm_val_loss, label='Val', color='green')
plt.plot(test_losses_lm, label='Test', color='red')
plt.scatter(np.argmin(lm_val_loss), min(lm_val_loss), color='green', s=30, zorder=5)
plt.axhline(y=min(lm_val_loss), linestyle='--', color='gray', linewidth=0.8)
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.title('Mean Squared Error - Levenberg-Marquardt')
plt.xlabel('Epochs')
plt.ylabel('MSE (log scale)')
plt.legend()

plt.tight_layout()
plt.show()

print(f'LM Lowest Validation Loss: {min(lm_val_loss)} at epoch {np.argmin(lm_val_loss)}' )


#Loss without value scaling
#LM Test Loss: 6.508686478667869e-09

# test size = 0.1
# learning rate = 0.5
```
